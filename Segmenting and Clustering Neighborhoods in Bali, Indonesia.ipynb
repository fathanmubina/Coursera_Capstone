# Segmenting and Clustering Neighborhoods in Bali, Indonesia

## Section 1: Retrieving Data

#### Importing libraries

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl
import re

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

print("Library Imported!")
```

#### Retrieving data from kodeposku.com

```python
#Entering URL
url = 'https://kodeposku.com/provinsi/bali'
html = urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, "html.parser")

# Retrieve all of the anchor tags
tags = soup('td')
list_tags = list()

#Creating a list for each column
postal_code = list()
village = list()
district = list()
regency = list()

#Creating a list of strings from tags
for tag in tags:
    # Look at the parts of a tag
    list_tags.append(str(tag))

#Finding the data for each column from tags    
for line in list_tags:
    #Finding postal code data
    w = re.findall('<td><a href="/kodepos/[0-9].*">([0-9].*)</a></td>',line)
    #Finding regency data
    z = re.findall('<td><a href="/daerah/bali/[a-z]+.*[a-z]+">(.*)</a></td>',line)
    
    #Creating a list of postal_code column
    for a in w:
        postal_code.append(a)
    #Creating a list of regency column
    for d in z:
        regency.append(d)

#Finding village or sub-district data
list_village = list()
j=1

while j < len(list_tags):
    list_village.append(list_tags[j])
    j=j+5

for vil in list_village:
    x = re.findall('<td><a href="/daerah/bali/.*/.*/.*/">(.*)</a></td>',vil)
    
    #Creating a list of village column
    for b in x:
        village.append(b)
        
#Finding district data
list_district = list()
i=2

while i < len(list_tags):
    list_district.append(list_tags[i])
    i=i+5

for dis in list_district:
    y= re.findall('<td><a href="/daerah/bali/.*">(.*)</a></td>',dis)

    #Creating a list of district column
    for c in y:
        district.append(c)
    
```

#### Creating a dataframe

```python
#Creating a dataframe
df_bali = pd.DataFrame()

df_bali["regency"] =regency
df_bali["district"] = district
df_bali["village"] = village
df_bali["postal_code"] = postal_code

#Checking the dataframe
df_bali.head(20)
```
